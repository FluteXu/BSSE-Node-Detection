```python
from lungmask import mask
from lungmask import utils
import SimpleITK as sitk
import numpy as np
import matplotlib.pyplot as plt
import os
import pydicom as dicom
import tqdm
import shutil
```

```python
# reference code
# https://github.com/JoHof/lungmask
```

```python
INPUT = '/home/flute/ShengNAS2/SharedProjectData/Luna16/images/1.3.6.1.4.1.14519.5.2.1.6279.6001.994459772950022352718462251777.mhd'
input_image = sitk.ReadImage(INPUT)
segmentation = mask.apply(input_image)  # default model is U-net(R231)
```

```python
# visualization
img = sitk.GetArrayFromImage(input_image)
for i in np.arange(img.shape[-3]):
    plt.imshow(img[i, :, :],cmap='gray')
    mask = np.ma.masked_where(segmentation[i,:,:]==0, segmentation[i,:,:])
    plt.imshow(mask, cmap='cool', alpha=0.2)
    plt.show()
```

```python
# save mask
result_image = sitk.GetImageFromArray(segmentation)
result_image.CopyInformation(input_image)
# write the image
sitk.WriteImage(result_image, 'lmask.nii')
```

## LIDC-DATASET-LUNG_MASK_GENERATION

```python
def clean_path(filepath):
    if not os.path.exists(filepath):
        os.makedirs(filepath)
    else:
        shutil.rmtree(filepath)
        os.makedirs(filepath)
```

```python
CTDir = '/home/flute/ShengNAS2/SharedProjectData/MICCAI_2022/LIDC-LIRI_flat'
save_dir = '/home/flute/ShengNAS2/SharedProjectData/MICCAI_2022/LIDC-LIRI_flat_lung_mask'
clean_path(save_dir)

for dirName, subdirList, fileList in os.walk(CTDir):
    for subdir in tqdm.tqdm(subdirList):
        if 'LIDC-IDRI' in subdir:
            pt_dcm_dir = os.path.join(CTDir, subdir)
            input_image = utils.get_input_image(pt_dcm_dir)
            
            segmentation = mask.apply(input_image)  # default model is U-net(R231)
            
            # save mask
            mask_save_pth = os.path.join(save_dir, subdir + '.nii')
            result_image = sitk.GetImageFromArray(segmentation)
            result_image.CopyInformation(input_image)
            # write the image
            sitk.WriteImage(result_image, mask_save_pth)
```
